#### NumPy:
*   `np.array()`: Создание массива из списка или кортежа.
*   `np.zeros()`, `np.ones()`: Создание массивов, заполненных нулями или единицами.
*   `np.arange()`: Создание массива с последовательными значениями.
*   `np.random.randint()`: Генерация случайных целых чисел.
*   `reshape()`: Изменение формы массива.
*   `max()`, `min()`, `mean()`: Вычисление максимального, минимального и среднего значений.
*   `.T`: Транспонирование массива (обмен строк и столбцов).
*   Арифметические операции (+, -, *, /): Поэлементные операции над массивами.
*   Индексация и срезы: Доступ к элементам или подмассивам.
*   Булева индексация: Выбор элементов по условию.
*   `.to_numpy()`: Преобразование DataFrame в массив NumPy.

#### Pandas:
*   `pd.DataFrame()`: Создание DataFrame из словаря, списка списков или других структур.
*   `pd.Series()`: Создание одномерной структуры данных Series.
*   `head()`: Просмотр первых строк DataFrame.
*   `info()`: Получение краткой сводки о DataFrame (типы данных, пропущенные значения).
*   `describe()`: Получение описательной статистики для числовых столбцов.
*   `df["колонка"]`: Выбор столбца по имени.
*   Булева индексация: Выбор строк по условию.
*   Добавление/удаление столбцов: Присваивание нового столбца или использование `drop()`.
*   `rename()`: Переименование столбцов.
*   `sum()`: Вычисление суммы элементов Series или столбца DataFrame.
*   `.str.upper()`: Применение строковых операций к элементам Series.
*   `fillna()`: Заполнение пропущенных значений.
*   `drop_duplicates()`: Удаление дублирующихся строк.
*   `pd.get_dummies()`: Преобразование категориальных данных в One-Hot Encoding.
*   `apply()`: Применение функции к Series или DataFrame.
*   `groupby()`: Группировка данных по одному или нескольким столбцам.
*   `np.linalg.det()`: Вычисление определителя матрицы (NumPy).

#### NumPy:
*   `isnull()`, `np.nan`: Проверка и представление пропущенных значений.
*   `np.mean()`, `np.median()`, `np.std()`: Статистические функции.
*   `np.sum(axis)`: Суммирование элементов по осям.
*   `np.power()`: Возведение в степень.
*   `np.clip()`: Ограничение значений массива.
*   Булева индексация: Удаление элементов.

#### Pandas:
*   `isnull()`, `sum()`: Проверка и подсчет пропущенных значений.
*   `fillna()`: Заполнение пропущенных значений (с различными стратегиями).
*   `drop_duplicates()`: Удаление дубликатов.
*   `astype()`: Изменение типа данных.
*   `mean()`, `median()`, `std()`, `mode()`: Статистические функции для Series/DataFrame.
*   `cumsum()`: Кумулятивная сумма.
*   `pd.to_datetime()`, `.dt.year`, `.dt.month`, `.dt.day`: Работа с датами и временем.
*   `.str.lower()`, `.str.contains()`, `.str.extract()`: Строковые операции.
*   `nunique()`: Количество уникальных значений.
*   `round()`: Округление значений.
*   `pct_change()`: Процентное изменение.
*   `pd.cut()`: Разделение данных на интервалы.
*   `corr()`: Вычисление корреляционной матрицы.
*   `groupby()`, `agg()`: Группировка и агрегация данных.
*   `idxmax()`: Индекс максимального значения.
